```python
import scipy
import numpy as np
import matplotlib.pyplot as plt
import itertools
from sklearn.decomposition import PCA, KernelPCA
from sklearn.metrics.pairwise import cosine_similarity, rbf_kernel
from PIL import Image as im
from sklearn import preprocessing
```

```python
def load_image(path):
    img = im.open(path)
    rgb_img = img.convert('RGB')
    points = []
    colours = []

    for x in range(rgb_img.width):
        for y in range(rgb_img.height):
            r, g, b = rgb_img.getpixel((x, y))
            hexc = '#%02x%02x%02x' % (r, g, b)
            if (r, g, b) != (255, 255, 255):
                point = [x, -y + img.height]
                points.append(point)
                colours.append(hexc)

    return points, colours
```

```python
points_first, colours_first = load_image("datasets-kernel/pca_benchmark_a.png")
points_second, colours_second = load_image("datasets-kernel/pca_benchmark_b.png")
```
